#### 第一部分 SQL应用

一、SQL介绍

SQL 结构查询语言(Structured Query Language) 是一个标准化的 广泛使用的语言，可以检索和更新关系表格和数据库中的数据。

二、SQL语法

CREATE语句 生成表; 

INSERT、DELETE语句插入和删除行;

UPDATE语句增加或修改在表的列里的数值; 

SELECT语句用来检索和操作存于表中的数据; 

（子句内的项用逗号分开）

三、语句实例

```sas
OPTIONS COMPRESS = YES;
```

```sas
/* CREATE */
PROC SQL;
CREATE TABLE TCUSTR(
    CUSTR_NBR VARCHAR(18),
    SEX INT
);
QUIT;

PROC PRINT DATA = TCUSTR;
RUN;
```

```sas
/* INSERT */
PROC SQL;
INSERT INTO 
TCUSTR(CUSTR_NBR, SEX)
VALUES("440101200109090011", 1)
VALUES("360101199901010012", 0)
;
QUIT;

PROC PRINT DATA = TCUSTR;
RUN;
```

```sas
/* DELETE */
PROC SQL;
DELETE FROM TCUSTR 
WHERE CUSTR_NBR = "440101200109090011";
QUIT;

PROC PRINT DATA = TCUSTR;
RUN;
```

```sas
/* UPDATE */
PROC SQL;
UPDATE TCUSTR 
SET SEX = 1 
WHERE CUSTR_NBR = "360101199901010012";
QUIT;

PROC PRINT DATA = TCUSTR;
RUN;
```

**其他用法**

ORDER BY:排序

GROUP BY:分组 

WHERE:筛选条件 

HAVING:筛选条件(分组后)

```sas
/* SELECT */
PROC SQL;
SELECT * 
FROM SASHELP.CARS(OBS = 5);
QUIT;

PROC SQL;
SELECT * 
FROM SASHELP.CARS 
WHERE MAKE = "Acura";
QUIT;

PROC SQL;
SELECT MAKE, MSRP
FROM SASHELP.CARS 
WHERE MAKE = "Acura";
QUIT;
```

```sas
/* SQL FUNC */
PROC SQL;
SELECT 
    COUNT(MSRP),
    SUM(MSRP),
    MAX(MSRP),
    MIN(MSRP),
    AVG(MSRP)
FROM SASHELP.CARS 
WHERE MAKE = "Acura";
QUIT;
```

```sas
PROC SQL;
SELECT 
    COUNT(MSRP) AS CNT_MSRP,
    SUM(MSRP) AS SUM_MSRP,
    MAX(MSRP) AS MAX_MSRP,
    MIN(MSRP) AS MIN_MSRP,
    AVG(MSRP) AS AVG_MSRP
FROM SASHELP.CARS 
WHERE MAKE = "Acura";
QUIT;
```

```sas
/* GROUP BY */
PROC SQL;
SELECT 
    MAKE,
    COUNT(MSRP) AS CNT_MSRP,
    SUM(MSRP) AS SUM_MSRP,
    MAX(MSRP) AS MAX_MSRP,
    MIN(MSRP) AS MIN_MSRP,
    AVG(MSRP) AS AVG_MSRP 
FROM SASHELP.CARS(OBS=100)
GROUP BY MAKE;
QUIT;
```

```sas
/*  ORDER BY*/
PROC SQL;
SELECT 
    MAKE,
    COUNT(MSRP) AS CNT_MSRP,
    SUM(MSRP) AS SUM_MSRP,
    MAX(MSRP) AS MAX_MSRP,
    MIN(MSRP) AS MIN_MSRP,
    AVG(MSRP) AS AVG_MSRP 
FROM SASHELP.CARS(OBS=100)
GROUP BY MAKE
ORDER BY CNT_MSRP;
QUIT;
```

```sas
/* WHERE */
PROC SQL;
SELECT 
    MAKE,
    COUNT(MSRP) AS CNT_MSRP,
    SUM(MSRP) AS SUM_MSRP,
    MAX(MSRP) AS MAX_MSRP,
    MIN(MSRP) AS MIN_MSRP,
    AVG(MSRP) AS AVG_MSRP 
FROM SASHELP.CARS(OBS=100)
WHERE MAKE ^= "Acura"
GROUP BY MAKE
ORDER BY CNT_MSRP;
QUIT;
```

```sas
/* HAVING */
PROC SQL;
SELECT 
    MAKE,
    COUNT(MSRP) AS CNT_MSRP,
    SUM(MSRP) AS SUM_MSRP,
    MAX(MSRP) AS MAX_MSRP,
    MIN(MSRP) AS MIN_MSRP,
    AVG(MSRP) AS AVG_MSRP 
FROM SASHELP.CARS(OBS=100) 
WHERE MAKE ^= "Acura"
GROUP BY MAKE
HAVING CNT_MSRP > 5
ORDER BY CNT_MSRP;
QUIT;
```

```sas
/* CREATE TABLE */
PROC SQL;
CREATE TABLE CARS_GROUP1 AS 
SELECT 
    MAKE,
    COUNT(MSRP) AS CNT_MSRP,
    SUM(MSRP) AS SUM_MSRP,
    MAX(MSRP) AS MAX_MSRP,
    MIN(MSRP) AS MIN_MSRP,
    AVG(MSRP) AS AVG_MSRP 
FROM SASHELP.CARS 
WHERE MAKE ^= "Acura"
GROUP BY MAKE
HAVING CNT_MSRP > 5
ORDER BY CNT_MSRP;
QUIT;

PROC PRINT DATA = CARS_GROUP1(OBS=10);
RUN;
```

```sas
/* CASE WHEN */
PROC SQL;
CREATE TABLE CARS_GROUP2 AS 
SELECT 
    MAKE,
    SUM(CASE WHEN ENGINESIZE > 3 THEN MSRP ELSE 0 END) AS SUM_MSRP
FROM SASHELP.CARS 
GROUP BY MAKE;
QUIT;

PROC PRINT DATA = CARS_GROUP2(OBS=10);
RUN;
```
